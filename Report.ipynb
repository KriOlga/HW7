# Краткий отчет о проекте по извлечению данных с сайта Wildberries

## URL сайта
Сайт: [Wildberries](https://www.wildberries.ru/)

https://www.wildberries.ru/

## Описание
Целью данного проекта было извлечение информации о товарах, в частности, названий товаров, цен со скидкой и URL-адресов товаров, с сайта Wildberries. Мы сосредоточились на категории "Камера видеонаблюдения", чтобы собрать данные о доступных моделях, их ценах и ссылках для дальнейшего изучения.

## Подход
Для навигации по сайту и извлечения данных использовался язык программирования Python с библиотеками Selenium и BeautifulSoup.
1. **Инициализация драйвера**: Мы использовали Selenium для автоматизации браузера Chrome, что позволило нам взаимодействовать с динамически загружаемыми элементами страницы.
2. **Поиск товаров**: После загрузки главной страницы мы нашли поле для поиска по ID и ввели запрос "Камера видеонаблюдения", инициировав поиск.
3. **Сбор данных**: Мы использовали цикл для прокрутки страницы вниз и ожидания загрузки карточек товаров. Затем с помощью BeautifulSoup извлекли нужные данные из HTML-кода, включая названия, цены со скидкой и URL-адреса товаров.
4. **Сохранение данных**: Собранные данные были сохранены в формате JSON и CSV для дальнейшего анализа.

## Трудности
В ходе реализации проекта возникли несколько проблем:
1. **Динамическая загрузка контента**: Некоторые элементы загружались асинхронно, что требовало ожидания их появления на странице. Это было решено с помощью WebDriverWait и ожидания элементов.
2. **Изменения в структуре HTML**: Структура страницы могла изменяться, что влияло на корректность селекторов. Мы проверяли текущую структуру HTML и адаптировали селекторы при необходимости.
3. **Обработка ошибок**: В процессе извлечения данных возникали ошибки, связанные с отсутствием некоторых элементов. Для их обработки были добавлены блоки try-except, что позволило избежать остановки программы из-за одной ошибки.

## Результаты
В результате работы был успешно собран набор данных, включающий названия товаров, цены со скидкой и URL-адреса на 20 товаров из категории "Камера видеонаблюдения". Данные были сохранены в файлы `products.json` и `products.csv`, что позволяет легко использовать их для дальнейшего анализа или визуализации. Проект продемонстрировал эффективность использования Selenium и BeautifulSoup для веб-скрейпинга и обработки данных.
